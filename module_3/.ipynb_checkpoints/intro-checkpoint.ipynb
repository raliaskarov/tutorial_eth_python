# Object Oriented Programming - Notes

```python

# demo class
class Person:
    def __init__(self, name, firstname, year_of_birth, job):
        self.name = name
        self.firstname = firstname
        self.year_of_birth = year_of_birth
        self.job = job
    def print(self):
        print(self.name + " " + self.firstname )
    def getAge(self, year):
        age = year - self.year_of_birth
        return age
```

```python
# create instance of a class Person (a.k.a. create Object)
p1 = Person(name = "Meier",
           firstname = "Maria",
           year_of_birth = 1954,
           job = "electrician")
#p2 = Person(); p3 = Person()
p1.name
```

```python
p1.print()
```
