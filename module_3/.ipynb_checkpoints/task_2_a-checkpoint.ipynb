```python
# Task
"""
Hotel
name: str
stars: int
floors: int
roomsPerFloor: int
occupancy: int
---------------------
printInfo(): none
getBookedRooms(): int
getMaxRooms(): int
checkin(): bool
checkout(): bool
"""
```

```python
class Hotel:
    def __init__ (self, name, stars, floors, roomsPerFloor, occupancy):
        """
            name: The name of the hotel is stored here.
            stars: Here is the number of hotel stars.
            floors: This indicates how many floors the hotel has.
            roomsPerFloor: This indicates how many rooms are on a floor.
            occupancy: This attribute stores how many rooms are currently occupied
        """
        self.name = name
        self.stars = stars
        self.floors = floors
        self.roomsPerFloor = roomsPerFloor
        self.occupancy = occupancy

    def printInfo(self):
        """
            Shows hotel name, number of stars, number of rooms, number of occupied rooms
        """
        name = self.name
        stars = self.stars
        total_rooms = self.getMaxRooms()
        booked_rooms = self.occupancy
        available_rooms = self.getBookedRooms()
        print(f"Hotel: {name}\nTotal rooms: {total_rooms}\nBooked rooms: {booked_rooms}\
                \nAvailale rooms: {available_rooms}")

    def getMaxRooms(self):
        """
            Total rooms in the hotel
        """
        totalRooms = self.roomsPerFloor * self.floors
        return totalRooms

    def getBookedRooms(self):
        """
            Description: How many rooms can be booked
            Input: floors, rooms per floor
            Output: number of bookable rooms
        """
        totalRooms = self.getMaxRooms()
        bookableRooms = totalRooms - self.occupancy
        return bookableRooms

    def checkin(self):
        """
            Increase occupancy by 1
        """
        total_rooms = self.getMaxRooms()
        if self.occupancy == total_rooms:
            print("Hotel is full")
            return False

        currentOccupancy = self.occupancy
        self.occupancy += 1
        print(f"Occupancy changed from {currentOccupancy} to {self.occupancy}")
        return True

    def checkout(self):
        """
            Decrease occupancy by 1
        """
        if self.occupancy == 0:
            print("Hotel is empty")
            return False
        currentOccupancy = self.occupancy
        self.occupancy -= 1
        print(f"Occupancy changed from {currentOccupancy} to {self.occupancy}")
        return True

    def copy(self):
        """
            Copy object
        """
        new_object = Hotel(self.name,
                           self.stars,
                           self.floors,
                           self.roomsPerFloor,
                           self.occupancy)
        return new_object

hotel_1 = Hotel(name = "Mariott",
                stars = 4,
                floors = 10,
                roomsPerFloor = 5,
                occupancy = 0)
# copy hotel
hotel_2 = hotel_1.copy()
hotel_2.name = "Hilton"
hotel_2.stars = 5
hotel_2.floors = 15
hotel_2.roomsPerFloor = 10
hotel_2.occupancy = 0

# list objects of class Hotel
hotels = [hotel_1, hotel_2]
for hotel in hotels:
    hotel.printInfo()
    print()


# book a room in hotel_1 10 times
for i in range (10):
  hotel_1.checkin()
hotel_1.printInfo()
```
