# Classes Demo

```python
# defin class 
class Animals:
    def __init__(self, name, genus, species, weight, length, speed, protection):
        self.name = name
        self.genus = genus
        self.species = species
        self.weight = weight
        self.length = length
        self.speed = speed
        self.protection = protection

    def printInfo(self):
        print(self.name)
        print(self.genus, self.species)
        print("weight: ", self.weight, "kg")
        print("length: ", self.length, "m")
        print(f"speed: {self.speed}")
        print(f"protection: {self.protection}")
        print(f"id: {id(self)}")

    def getTime(self, distance):
        if distance > 0:
            time = distance / (self.speed / 3.6)
            time = round(time, 2)
            return time
        else:
            return 0

    def feed(self, food_weight):
        print(f"Weight before: {self.weight} kg")
        if food_weight >= 0:
            self.weight += food_weight*0.3
            print(f"Weight after: {self.weight}")
        else:
            print("Invalid value.")


    def copy(self):
        new_object = Animals(self.name,
                             self.genus,
                             self.species,
                             self.weight,
                             self.length,
                             self.speed,
                             self.protection)
        return new_object

an1 = Animals("blue whale", "Balaenoptera" , "musculus", 200000, 32.0, 48, True)
an2 = Animals("big dolphin", "Tursiops",  "truncatus", 450, 6.5, 60, False)
an3 = Animals("bluefin tuna", "Thunnus", "thynnus", 300, 3.2, 70, False)
an4 = Animals("swordfish", "Xiphias", "gladius", 130, 1.5, 97, False)
an5 = Animals("green sea turtle", "Chelonia", "mydas", 165, 1.1, 0.5, True)
an1.printInfo()
distance = 50
print(f"Time to cover distance {distance}m: {an1.getTime(distance)}")
an1.feed(100)
```
