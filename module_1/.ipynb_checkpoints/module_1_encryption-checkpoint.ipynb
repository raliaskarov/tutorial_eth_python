### Answer

````python
# DNA sequence to test
test_seq = "TTGTCTTCTAAATAAAGACCAGTATGAACGGCTAAATGAGGACAAACCTGTCTCTTATAATTAATAAGTGAAACTGATCTCCCAGTACAAAAGCTGGGATATTAACATAAGACGAGAAGACCCCGTGGAGCTTGAATTAAACCACTAAACAATACCTAGTTTATGTACCATTTCCAGTTGGGGCGACTTCGGAACAAAACAAAACTTCCGAGCACAGAACCACTCCTCTAACCAAGGCCAACCCTTAGGCTGGGGATCGACCCGGCCCTTGCCGATCAACGAACCAAGTTACCCCGGGGATAACAGCGCTATCTTCTTCAAGAGTCCCTATCAACAAGAAGGTTTACGACCTCGATGTTGGATCAGGACACCCAAATGGTGCAGCCGCTATTAAAGGTTCGTTTGTTCAACGATT"

# DNA sequences to compare the test sequences against
seq_nontoxic  ="TTGTCTTCTAAATAAAGACCTGTATGAACGGCTAAATGAGGATAAACCTGTCTCTTATAACCAACCAGTGAAACTGATCTCCCAGTACAAAAGCTGGAATCAACCCATAAGACGAGAAGACCCTGCGGAGCTTAAAACAACCACTAATTAATATCTAACCCTATGAAATGTTTTCAGTTGGGGCGACTTCGGAATAAAACACAACTTCCGAGCACAGAACCACAATTCTAACCAAGGCCAACAAGCCAAAGTATATATGACCCGGCACATGCCGATCAACGAACCAAGTTACCCCAGGGATAACAGCGCTATCCTCTTCAAGAGTCCCTATCAACAAGAGGGTTTACGACCTCGATGTTGGATCAGGACACCCAAATGGTGCAGCCGCTATTAAAGGTTCGTTTGTTCAACGATT"
seq_toxic     ="TTGTCTTCTAAATAAAGACCAGTATGAACGGCTAAATGAGGACAAACCTGTCTCTTATAATTAATCAGTGAAACTGATCTCCCAGTACAAAAGCTGGGATATTAACATAAGACGAGAAGACCCCGTGGAGCTTGAAATAAACCACTAAACAATACCTAGTTTATGTACTATTTTCAGTTGGGGCGACTTCGGAACAAAACAAAACTTCCGAGCACAGAACCACTCCTCTAACCAAGGCCAACAAGCCTAAGACTCATCGACCCGGCCCTTGCCGATCAACGAACCAAGTTACCCCGGGGATAACAGCGCTATCTTCTTCAAGAGTCCCTATCAACAAGAAGGTTTACGACCTCGATGTTGGATCAGGACACCCAAATGGTGCAGCCGCTATTAAAGGTTCGTTTGTTCAACGATT"

# -------------------------------------------------------
# Part 1 - Analyze DNA (count occurrences)
# -------------------------------------------------------
bases = ["A", "T", "C", "G"]
occurrences = [0] * len(bases)

for i in range(len(bases)):           # loop over each base type
    count = 0
    for j in range(len(test_seq)):    # loop over each character in DNA
        if test_seq[j] == bases[i]:
            count += 1
    occurrences[i] = count

print("Occurrences of Bases:")
for i in range(len(bases)):
    print(bases[i], ": ", occurrences[i], sep="")
print()

# -------------------------------------------------------
# Part 2 - Determine species (toxic or nontoxic)
# -------------------------------------------------------
diff_nontoxic = 0
diff_toxic = 0

for i in range(len(test_seq)):
    if test_seq[i] != seq_nontoxic[i]:
        diff_nontoxic += 1
    if test_seq[i] != seq_toxic[i]:
        diff_toxic += 1

print("The test sample is probably", end=" ")
if diff_nontoxic < diff_toxic:
    print("nontoxic")
else:
    print("toxic")
print()

# -------------------------------------------------------
# Part 3 - Longest identical subsequence
# -------------------------------------------------------
def longest_common_substring(s1, s2):
    # DP table for LCS
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]
    longest = 0
    end_index = 0
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
                if dp[i][j] > longest:
                    longest = dp[i][j]
                    end_index = i
    start_index = end_index - longest
    return start_index, longest

start_nt, len_nt = longest_common_substring(test_seq, seq_nontoxic)
start_t, len_t  = longest_common_substring(test_seq, seq_toxic)

if len_nt > len_t:
    print("Longest identical subsequence with: nontoxic")
    print("start:", start_nt)
    print("length:", len_nt)
else:
    print("Longest identical subsequence with: toxic")
    print("start:", start_t)
    print("length:", len_t)
````

Part 3 - The Vigenère entcrypton

Vigenère encryption is an extension of Caesar encryption that offers more security: While Caesar encryption shifts each letter in the plaintext by a fixed number of positions in the alphabet, Vigenère encryption uses an entire keyword or key phrase. Each letter in the plaintext is shifted by the corresponding letter in the keyword. The shift is thus still performed in the alphabet, but it changes for each letter based on its position in the keyword.

Example of Vigenère entcryption

Plain text: INFORMATICS
Key: "ABCD"
Caesar ciphertext: IOHRRNCWIDU

As keys are often shorter than plain text, you simply start again at the first position of the key when you reach the end of the key.

### Answer

````python
# ================ Part 1 ================
# Complete the following function
# The function offset_character has two parameters:
# - char: contains a single letter
# - offset: contains an integer
# The function should move the letter in char to the left by offset positions 
# and return the result.

def offset_character(char, offset):
    # Work only on letters
    if char.isalpha():
        # Keep uppercase as uppercase, lowercase as lowercase
        base = ord('A') if char.isupper() else ord('a')
        # Shift left by offset (mod 26 ensures wrap-around)
        shifted = (ord(char) - base - offset) % 26 + base
        return chr(shifted)
    else:
        # Non-letters are unchanged
        return char


# ================ Part 2 ================
# Decrypt the text stored in the variable ciphertext_caesar.
# Pay attention to special characters!

ciphertext_caesar = "GUR PNRFNE PVCURE JVGU GUR SVKRQ FUVSG BS 13 CYNPRF VF NYFB XABJA NF GUR EBG13 FUVSG. VG JNF IREL CBCHYNE VA GUR 80F NAQ JNF HFRQ VA INEVBHF SBEHZF GB CEBGRPG ERNQREF (R.T. SEBZ NPPVQRAGNYYL ERNQVAT FCBVYREF BE CHAPUYVARF)."
key = 13

plaintext_caesar = ""

for ch in ciphertext_caesar:
    if ch.isalpha():
        plaintext_caesar += offset_character(ch, key)
    else:
        plaintext_caesar += ch

print("=========== CAESAR DECRYPTION ===========")
print("Ciphertext:")
print(ciphertext_caesar)
print("Key:", key)
print("Plaintext:")
print(plaintext_caesar)
print()


# ================ Part 3 ================
# Decrypt the text stored in the variable ciphertext_vigenere.
# Pay attention to special characters!

ciphertext_vigenere = "AUTX REQ RBKYD JITL RXH IVZLK IWIF? VSKFX IMIKXCW."
key = "ENTERKEY"

plaintext_vigenere = ""

key_index = 0
for ch in ciphertext_vigenere:
    if ch.isalpha():
        # Find correct key letter
        key_char = key[key_index % len(key)]
        # Convert key letter to shift value (A=0, B=1, ..., Z=25)
        shift = ord(key_char.upper()) - ord('A')
        # Shift the ciphertext character left by this amount
        plaintext_vigenere += offset_character(ch, shift)
        key_index += 1  # advance key only on letters
    else:
        plaintext_vigenere += ch

print("========== VIGENERE DECRYPTION ==========")
print("Ciphertext:")
print(ciphertext_vigenere)
print("Key:", key)
print("Plaintext:")
print(plaintext_vigenere)
print()
````
