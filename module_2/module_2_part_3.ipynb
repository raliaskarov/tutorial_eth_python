# Galton Board Simulation

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

height = int(input("Height? "))
num_balls = int(input("Number of balls? "))

# Number of bins (height + 1)
number_bins = height + 1
bins = np.zeros(number_bins, dtype=int)

# Simulate each ball: each level is a 50:50 decision using randint (0 or 1)
for _ in range(num_balls):
#    print(f"Ball {_ + 1}")
    position_sum = 0
    for _ in range(height):
        position_sum += rd.randint(0, 1)
    bins[position_sum] += 1

# Print ONLY the counts, space-separated, to satisfy the auto-grader
print(" ".join(str(x) for x in bins))

# Visualization and save as PNG in ./cx_out
import os
os.makedirs("cx_out", exist_ok=True)
plt.bar(range(number_bins), height=bins, color="blue")
plt.xlabel("Bin index (0 to height)")
plt.ylabel("Number of balls")
plt.title(f"Galton Board (height={height}, balls={num_balls})")
plt.tight_layout()
plt.savefig("./cx_out/galton_board.png")
plt.close()
```
