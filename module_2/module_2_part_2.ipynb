## 02 / Case Study I: Heat propagation in a metal rod (1d-model) - Student attempt

We have a metal rod , which has a temperature of 10 ° Celsius along its entire length. Now we hold a heat source (e.g. a burning candle) at both ends at different distances under the stick. This creates a constant heat source of 40 ° C on the left side , and one of 100 ° C on the right end of the bar .

 order to be able to calculate the heat propagation, we have to discretize the metal rod, i.e. break it down into several parts:

    The metal rod is divided into 9 compartments (cells).
    We assume that within two measurement time points 40% of the temperature difference is passed on between two neighboring cells, with the heat flowing from the warmer to the colder cell .
    We assume that the two heated end cells keep their temperature constant .

The metal rod divided into cells can be interpreted as a vector . The individual cells of the rod correspond to the elements of the vector.

When developing this simulation, we proceed as follows:

    Initial situation of the simulation (steps 2 - 6)
    Calculations of the model (steps 7-11)
    Visualization of the results (steps 12 - 13)


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import os

# ensure dirs
os.makedirs("./cx_out", exist_ok=True)

# def vector
v = np.array([40] + [10]*8 + [100])
prop = 0.4 
n_cycles = 12


#variables
cells = v.size
prop=0.4


# 2nd vetor
v_old = np.copy(v)

#v_new = np.zeros(cells, int)
#v_new[0] = v_old[0]
#v_new[cells-1] = v_old[cells-1]
#print(v_new)

#plot
fig, ax = plt.subplots(figsize = (6, 4))
bars = ax.bar(range(cells), v,
             color="blue")
ax.set(title=f"Heat propagation in a metal rod",
       xlabel = "Cell",
       ylabel = "Temperature, C",
      )
#ax.bar_label(bars, padding = -15, color = 'white')
#plt.savefig("cx_out/case_study_metal_rod.png")


def animate(cycle):
    global v_old
    v_new = v_old.copy()
    for i in range (1, cells -1):
        v_new[i] = (1 - 2 * prop) * v_old[i] + prop * (v_old[i-1] + v_old[i+1])
        v_old = v_new.copy()
    print(f"v_new: {v_new}")

    # update bars
    for bar, h in zip(bars, v_new):
        bar.set_height(h)
    # remove old labels and add new ones
    [lbl.remove() for lbl in ax.texts]
    ax.bar_label(bars, padding=-15, color="white", fmt="%.0f")

    # update title
    ax.set_title(f"Heat propagation in a metal rod (Cycle: {cycle})")

    v_old = v_new
    return bars
    
for cycle in range (0, n_cycles):
    animate(cycle)


ani = animation.FuncAnimation(fig, animate, cycle+1, init_func=lambda *args: None)
writer = animation.PillowWriter(fps=1)
ani.save("./cx_out/case_study_metallstab.gif", writer=writer)
```

## Case Study II Matrix heat propagation model

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

# ----------------------------
# 1) Setup: matrix + sources
# ----------------------------
# Use int for exact match with tutorial; convert to float for math
plate = np.zeros((24, 24), dtype=int)
plate[:, :] = 20

# Fixed heater lines (1 cell wide)
def apply_sources(M):
    M[4, 4:21]  = 99
    M[5:10, 20] = 80
    M[11:20, 4] = 50
    M[20, 4:21] = 60

apply_sources(plate)

# Simulation parameters
prop   = 0.2
cycles = 50
H, W   = plate.shape

# Working copies as float to avoid truncation during arithmetic
plate_old = plate.astype(float, copy=True)
plate_new = plate.astype(float, copy=True)

# Border condition: always ambient 20 °C
def apply_border(M):
    M[0,  :] = 20
    M[-1, :] = 20
    M[:,  0] = 20
    M[:, -1] = 20

apply_border(plate_old)  # enforce at start as well

# ----------------------------
# 2) Visualization scaffolding
# ----------------------------
fig = plt.figure()
contour = plt.contourf(plate_old)
cbar = plt.colorbar(contour)

# ----------------------------
# 3) Animate: one diffusion step
# ----------------------------
def animate(step):
    global plate_old, plate_new  # tutorial asks for global to write them

    # Update interior using 4-neighbor stencil
    for i in range(1, H-1):
        for j in range(1, W-1):
            plate_new[i, j] = ((1 - 4*prop) * plate_old[i, j]
                               + prop * (plate_old[i-1, j] + plate_old[i+1, j]
                                         + plate_old[i, j-1] + plate_old[i, j+1]))

    # Re-apply constraints every step
    apply_border(plate_new)
    apply_sources(plate_new)

    # Redraw
    plt.cla()
    plt.title(f"Heat propagation in a plate (cycle: {step})")
    cont = plt.contourf(plate_new)
    cbar.update_normal(cont)

    # Advance: old <- new
    plate_old[:, :] = plate_new[:, :]

# ----------------------------
# 4) Run + save animation
# ----------------------------
ani = animation.FuncAnimation(
    fig, animate, frames=cycles, interval=10, blit=False
)

# Save GIF (1 frame per second)
writer = animation.PillowWriter(fps=1)
ani.save("cx_out/case_study_plate.gif", writer=writer)

# Optional: also save final frame as PNG
plt.figure()
plt.title("Final temperature field")
plt.contourf(plate_old)
plt.colorbar()
plt.savefig("cx_out/case_study_plate_final.png", dpi=150)
```
