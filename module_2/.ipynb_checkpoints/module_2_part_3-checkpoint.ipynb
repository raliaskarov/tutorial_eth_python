# Galton Board Simulation

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

height = int(input("Height? "))
num_balls = int(input("Number of balls? "))

# Number of bins for a Galton board of given height is height + 1
bins = np.zeros(height + 1, dtype=int)

# Simulate each ball
for _ in range(num_balls):
    # Position equals the number of right moves (1s) among "height" 50:50 decisions
    position_sum = 0
    for _ in range(height):
        # 0 for left, 1 for right (50:50 using random.random)
        step = 1 if rd.random() >= 0.5 else 0
        position_sum += step
    # position_sum is already the correct bin index (0..height)
    bins[position_sum] += 1

# Print the bin counts in the required format
print("Number of balls in the bins:")
print(" ".join(str(x) for x in bins))

# Plot and save as PNG (required by tests)
fig, ax = plt.subplots()
ax.bar(range(height + 1), bins)
ax.set_xlabel("Bin index (0 to height)")
ax.set_ylabel("Number of balls")
ax.set_title(f"Galton Board (height={height}, balls={num_balls})")
plt.tight_layout()
plt.savefig("galton_board.png")
plt.close(fig)
```
