```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import os
```

```python
def describe_array(array):
    print(f"Type: {type(array)}")
    print(f"Data type: {array.dtype.name}")
    print(f"Size: {array.size}")
    print(f"Dimensions: {array.ndim}")

def show_stats(array):
    print("N=", np.size(array))
    print("Min=",np.min(array))
    print("Max=",np.max(array))
    print("Median=",np.median(array))
    print("Mean=",np.mean(array))
    print("Std=",np.std(array))
```

```python
# test csv with 2 rows
try:
    with open("test.csv", 'r') as csvfile:
        data = list(csv.reader(csvfile, delimiter=","))
        print(f"data: {data}")
        print(type(data))
        test_array = np.array(data, float)
        test_array_T = np.array(data, float).T
        print(f"\ntest_array:\n{test_array}")
        describe_array(test_array)
        print(f"\ntest_array_t:\n{test_array_T}")
        describe_array(test_array_T)
except FileNotFoundError:
    pass
```

```python
with open("ozonzh2020.csv", 'r') as csvfile:
    data = list(csv.reader(csvfile, delimiter=","))
ozone = np.array(data, float).T
```
